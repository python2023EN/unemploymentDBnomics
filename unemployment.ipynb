```python
#load US and EU unemployment data from OECD
usa_unem = fetch_series('OECD/ALFS_SUMTAB/USA.YGTT06PC_ST.A')
euro_unem = fetch_series('OECD/ALFS_SUMTAB/EA19.YGTT06PC_ST.A')
```

```python
#filter data
usa_subset = usa_unem[usa_unem['period'].ge('2005-01-01')]
```

```python
#filter data
euro_subset = euro_unem[euro_unem['period'].ge('2005-01-01')]
```

```python
#verify that both subsets are with same shape
print(euro_subset.shape)
print(usa_subset.shape)
```

```python
#merge the 2 subsets
useu = pd.concat([euro_subset, usa_subset], ignore_index=True, axis=0)
```
